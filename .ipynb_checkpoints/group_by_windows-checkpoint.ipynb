```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
from sklearn.decomposition import PCA
from sklearn.covariance import EllipticEnvelope
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
%matplotlib inline
```

```python
df = pd.read_csv("dataset.csv")
df.describe()
```

```python
df_windows_median = df.groupby('day_cycle_window').median()
df_windows_std = df.groupby('day_cycle_window').std()
```

```python
df[df['day_cycle_window'] == '1_1_1'].std()
```

```python
std = pd.DataFrame()
median = pd.DataFrame()
for w in df['day_cycle_window'].unique():
    std.append(df[df['day_cycle_window'] == w].std(), ignore_index = True)
    median.append(df[df['day_cycle_window'] == w].median(), ignore_index = True)
    
    
std['day_cycle_window'] = df['day_cycle_window'].unique()
median['day_cycle_window'] = df['day_cycle_window'].unique()
```
